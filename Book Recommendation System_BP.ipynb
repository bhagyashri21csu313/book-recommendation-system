```python
ratings_with_books=ratings.merge(books,on='ISBN')
```

```python
ratings_with_books
```

```python
number_rating=ratings_with_books.groupby('Book-Title')['Book-Rating'].count().reset_index()
```

```python
number_rating.rename(columns={'Book-Rating':'number of ratings'},inplace=True)
```

```python
final_rating=ratings_with_books.merge(number_rating,on='Book-Title')
```

```python
final_rating.head()
```

```python
final_rating=final_rating[final_rating['number of ratings']>=50]
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns='User-ID',index='Book-Title',values='Book-Rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances,suggestions=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
book_pivot.index[54]
```

```python
def recommend_book(book_name):
    bookid=np.where(book_pivot.index==book_name)[0][0]
    distances,suggestions=model.kneighbors(book_pivot.iloc[bookid,:].values.reshape(1,-1),n_neighbors=6)
    
    for i in range(len(suggestions)):
        if i==0:
            print("Recommended Books Are :")
        if not i:
            print(book_pivot.index[suggestions[i]])
```

```python
recommend_book('Animal Farm')
```
